```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

lab2_data = sns.load_dataset('titanic')
lab2_data
```

```python
sns.barplot(data=lab2_data, x='class', y='survived')
plt.show()
```

```python
sns.barplot(data=lab2_data, x='class', y='survived', hue='who')
plt.show()
```

```python
lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'male'), 'who'] = 'boy'
lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'female'), 'who'] = 'girl'
sns.barplot(data=lab2_data, x='class', y='survived', hue='who')
plt.show()
```

```python
sns.barplot(data=lab2_data, x='class', y='survived', hue='who', ci=90, capsize=0.05)
plt.show()
```

```python
sns.barplot(data=lab2_data, x='survived', y='age', hue='who')
plt.show()
```

```python
lab2_data['survived_str'] = lab2_data['survived'].map({0: 'no', 1: 'yes'})
sns.barplot(data=lab2_data, x='survived_str', y='age', hue='who')
plt.show()
```

```python
sns.barplot(data=lab2_data, x='survived_str', y='age', hue='who', estimator=np.median, ci='sd', capsize=0.05)
plt.show()
```

```python
sns.boxplot(data=lab2_data, x='survived_str', y='age', hue='who')
plt.show()
```

```python
sns.stripplot(data=lab2_data, x='survived_str', y='age', hue='who', jitter=0.3, alpha=0.5, dodge=True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

lab2_data = sns.load_dataset('titanic')

plt.figure(figsize=(8, 6))
sns.barplot(data=lab2_data, x='class', y='survived')
plt.title('Survival Rate by Class')
plt.savefig('/content/practiceRepository/figure/Figure1.png')

plt.figure(figsize=(8, 6))
sns.barplot(data=lab2_data, x='class', y='survived', hue='who')
plt.title('Survival Rate by Class and Who')
plt.savefig('/content/practiceRepository/figure/Figure2a.png')

lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'male'), 'who'] = 'boy'
lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'female'), 'who'] = 'girl'

plt.figure(figsize=(8, 6))
sns.barplot(data=lab2_data, x='class', y='survived', hue='who')
plt.title('Survival Rate by Class and Who (Including Boys and Girls)')
plt.savefig('/content/practiceRepository/figure/Figure2b.png')

plt.figure(figsize=(8, 6))
sns.barplot(data=lab2_data, x='survived', y='age', hue='who')
plt.title('Age of Survivors and Non-Survivors by Who')
plt.savefig('/content/practiceRepository/figure/Figure3.png')

lab2_data['survived_str'] = lab2_data['survived'].map({0: 'no', 1: 'yes'})

plt.figure(figsize=(8, 6))
sns.barplot(data=lab2_data, x='survived_str', y='age', hue='who', estimator=np.median, ci='sd', capsize=0.05)
plt.title('Age of Survivors and Non-Survivors (Median, SD)')
plt.savefig('/content/practiceRepository/figure/Figure4.png')
```
